```python
path = "d:/teti/aulas/TETI_ATV4/imagens/"
# Definir imagens
imagens = [
    'badger.jpg',
    'badger2.jpg',
    'badger3.jpg'
]
```

```python
import matplotlib.pyplot as plt
from PIL import Image


# Exibir imagens
for imagem in imagens:
    img_path = path + imagem  # Concatenar o caminho com o nome da imagem
    img = Image.open(img_path)
    
    # Exibir imagem
    plt.imshow(img)
    plt.axis('off')  # Remover os eixos
    plt.show()
```

```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.applications.mobilenet import MobileNet
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np

path = "d:/teti/aulas/TETI_ATV4/imagens/"
# Definir modelos
modelos = [
    (ResNet50(weights='imagenet'), 'ResNet50'), #defini com o nome para exibir depois
    (VGG16(weights='imagenet'), 'VGG16'),
    (MobileNet(weights='imagenet'), 'MobileNet')
]

# Loop para testar cada modelo com cada imagem
for modelo, nome_modelo in modelos:
    for imagem in imagens:
        # Carregar imagem
        img_path = path + imagem  # Concatenar o caminho com o nome da imagem
        img = image.load_img(img_path, target_size=(224, 224))
        x = image.img_to_array(img)
        x = np.expand_dims(x, axis=0)
        x = preprocess_input(x)

        # Obter as previses
        preds = modelo.predict(x)
        previsoes = decode_predictions(preds, top=3)[0]

        # Exibir resultados
        print(f'Previses para {imagem} usando {nome_modelo}:')
        for _, classe, probabilidade in previsoes:
            print(f'{classe}: {probabilidade:.2%}')
        print()

        
        
#skunk = gamba
#guenon = macaco branco e preto
#groenendael= um tipo de pastor belga
#weasel, mink, polecat = bichos estilo doninha, furo, os mustela em geral
#porcupine = porco espinho
#marmot = marmota, parece um esquilo gigante

#o mobilenetv2 fez previsoes ruins, confundiu com tecidos e coisas do tipo
```

```python
#IMAGEM 1

from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.inception_v3 import preprocess_input, decode_predictions
import numpy as np
import matplotlib.pyplot as plt

modelo = InceptionV3(weights='imagenet')

img = image.load_img('d:/teti/aulas/TETI_ATV4/imagens/badger.jpg', target_size = (299, 299)) #TIVE QUE MUDAR DE 224 PRA 299

x = image.img_to_array(img)
x = np.expand_dims(x, axis = 0)
x = preprocess_input(x)

preds = modelo.predict(x)

# Obter as previses
previsoes = decode_predictions(preds, top=5)[0]

print('\nPrevises:\n')
for _, classe, probabilidade in previsoes:
    print(f'{classe}: {probabilidade:.20f}\n')
```
